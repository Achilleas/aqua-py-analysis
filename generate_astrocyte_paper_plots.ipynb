```python
single_multi_stick_path = '/Users/achilleasgeorgiou/Desktop/data_output/single_multi_stick'
```

```python

multi_stick_pairs = datasets['dataset_2']['pair']['single_to_multi_stick_pairs_L']
figs = {}
for measure_id in ['dffMax2', 'area', 'time_s']:
    mean_running_l = []
    mean_single_l = []
    mean_multi_l = []
    for multi_stick_pair in multi_stick_pairs:
        single_a, multi_a = multi_stick_pair['single'], multi_stick_pair['multi']
        print(single_a.base_path, single_a.day)

        mean_single_rest = np.mean(single_a.res_d[measure_id][single_a.event_subsets['rest']])
        mean_single_running_no_stick = np.mean(single_a.res_d[measure_id][single_a.event_subsets['running_no_stick']])
        mean_single_stick = np.mean(single_a.res_d[measure_id][single_a.event_subsets['stick_run_ind_15']])

        mean_multi_rest = np.mean(multi_a.res_d[measure_id][multi_a.event_subsets['rest']])
        #Stick everywhere, use running
        mean_multi_stick = np.mean(multi_a.res_d[measure_id][multi_a.event_subsets['running']])
        
        mean_running = mean_single_running_no_stick / mean_single_rest
        mean_single = mean_single_stick / mean_single_rest
        mean_multi = mean_multi_stick / mean_multi_rest

        mean_running_l.append(mean_running)
        mean_single_l.append(mean_single)
        mean_multi_l.append(mean_multi)

        # Create bar chart here
        print(measure_id, f'Running : {mean_running} Single {mean_single} Multi {mean_multi}')

    conf_running = np.mean(mean_running_l) - stat_utils.mean_confidence_interval(mean_running_l, confidence=0.95)[1]
    conf_single = np.mean(mean_single_l) - stat_utils.mean_confidence_interval(mean_single_l, confidence=0.95)[1]
    conf_multi = np.mean(mean_multi_l) - stat_utils.mean_confidence_interval(mean_multi_l, confidence=0.95)[1]

    mean_running = np.mean(mean_running_l)
    mean_single = np.mean(mean_single_l)
    mean_multi = np.mean(mean_multi_l)
    
    d = {
        'mean_running' : [mean_running],
        'conf_running' : [conf_running],
        'mean_single' : [mean_single],
        'conf_single' : [conf_single],
        'mean_multi' : [mean_multi],
        'conf_multi' : [conf_multi]
    }
    print('SAVING', os.path.join(single_multi_stick_path, 'plots', 'bar_comparison', measure_id + '.csv'))
    saving_utils.dict_to_csv(d, measure_id + '.csv', base_folder=os.path.join(single_multi_stick_path, 'plots', 'bar_comparison'))

    d_data = {
        'running' : mean_running_l,
        'single' : mean_single_l,
        'multi' : mean_multi_l
    }
    saving_utils.dict_to_csv(d_data, measure_id + '-data.csv', base_folder=os.path.join(single_multi_stick_path, 'plots', 'bar_comparison'))
    figs[measure_id] = plotly_utils.plot_bar(['running', 'single_stick', 'multi_stick'], [np.mean(mean_running_l), np.mean(mean_single_l), np.mean(mean_multi_l)], text_values=None, title='', text_size=10, x_title=measure_id, y_title='', margin_l=100, margin_b=100, err_y=[conf_running, conf_single, conf_multi], err_symmetric=True)

mean_running_l = []
mean_single_l = []
mean_multi_l = []

for multi_stick_pair in multi_stick_pairs:
    single_a, multi_a = multi_stick_pair['single'], multi_stick_pair['multi']
    print(single_a.base_path, single_a.day)
    #Number of events running / num frames running * 1 min of frames to get number of events running per minute of running
    single_events_running = len(single_a.event_subsets['running_no_stick'])  / len(single_a.indices_d['running_no_stick']) * single_a.minute_frames
    single_events_rest = len(single_a.event_subsets['rest'])  / len(single_a.indices_d['rest']) * single_a.minute_frames

    single_events_stick =  len(single_a.event_subsets['stick_run_ind_15'])  / len(single_a.indices_d['stick_run_ind_15']) * single_a.minute_frames

    multi_events_stick = len(multi_a.event_subsets['running'])  / len(multi_a.indices_d['running']) * multi_a.minute_frames
    multi_events_rest = len(multi_a.event_subsets['rest'])  / len(multi_a.indices_d['rest']) * multi_a.minute_frames

    running_events = single_events_running / single_events_rest
    single_events = single_events_stick / single_events_rest
    multi_events = multi_events_stick / multi_events_rest

    print(measure_id, f'Running events: {running_events} Single events{single_events} Multi events {multi_events}')
    mean_running_l.append(running_events)
    mean_single_l.append(single_events)
    mean_multi_l.append(multi_events)


conf_running = np.mean(mean_running_l) - stat_utils.mean_confidence_interval(mean_running_l, confidence=0.95)[1]
conf_single = np.mean(mean_single_l) - stat_utils.mean_confidence_interval(mean_single_l, confidence=0.95)[1]
conf_multi = np.mean(mean_multi_l) - stat_utils.mean_confidence_interval(mean_multi_l, confidence=0.95)[1]

mean_running = np.mean(mean_running_l)
mean_single = np.mean(mean_single_l)
mean_multi = np.mean(mean_multi_l)

d = {
    'mean_running' : [mean_running],
    'conf_running' : [conf_running],
    'mean_single' : [mean_single],
    'conf_single' : [conf_single],
    'mean_multi' : [mean_multi],
    'conf_multi' : [conf_multi]
}

saving_utils.dict_to_csv(d, 'count' + '.csv', base_folder=os.path.join(single_multi_stick_path, 'plots', 'bar_comparison'))

d_data = {
    'running' : mean_running_l,
    'single' : mean_single_l,
    'multi' : mean_multi_l
}

saving_utils.dict_to_csv(d_data, 'count' + '-data.csv', base_folder=os.path.join(single_multi_stick_path, 'plots', 'bar_comparison'))

figs['count'] = plotly_utils.plot_bar(['running', 'single_stick', 'multi_stick'], [np.mean(mean_running_l), np.mean(mean_single_l), np.mean(mean_multi_l)], text_values=None, title='', text_size=10, x_title='Events', y_title='', margin_l=100, margin_b=100, err_y=[conf_running, conf_single, conf_multi], err_symmetric=True)

for measure_id in figs.keys():
    path = os.path.join(single_multi_stick_path, 'plots', 'bar_comparison', measure_id)
    saving_utils.save_plotly_fig(figs[measure_id], path)
```
